# Intro to Seaborn

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

## Scatterplots

```python
df = pd.read_csv('../finished-notebooks/05-Seaborn/dm_office_sales.csv')
df.head()
```

```python
sns.scatterplot(x=df['salary'], y=df['sales'], hue=df['level of education'], palette='Dark2')
```

## Distribution Plots

```python
df.head()
```

```python
sns.rugplot(x=df['salary'], height=0.5)
```

```python
# DO NOT USE DISTPLOT--- will be deprecated
sns.set_theme(style='whitegrid')
sns.displot(x=df['salary'], bins=20, kde=True)
```

```python
sns.histplot(x=df['salary'])
```

```python
sns.kdeplot(x=df['salary'])
```

```python
np.random.seed(42)
sample_ages = np.random.randint(0, 100, 200)
sample_ages
```

```python
sample_ages = pd.DataFrame(sample_ages, columns=['age'])
```

```python
sample_ages
```

```python
sns.rugplot(x=sample_ages['age'], height=0.5)
```

```python
sns.displot(x=sample_ages['age'], bins=30, kde=True)
```

```python
sns.kdeplot(x=sample_ages['age'], clip=[0, 100], bw_adjust=0.6, fill=True)
```
