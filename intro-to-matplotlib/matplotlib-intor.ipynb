# Intro to MatplotLib

## OOP Figure Plotting Using Axes

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data
a = np.linspace(0, 10, 11)
b = a**4
```

```python
x = np.arange(0, 10)
y = 2 * x
```

```python
fig, axs= plt.subplots(1, 2, figsize=(10, 5))
axs[0].plot(a, b)
axs[1].plot(x, y)
axs[0].set_title('title is here')
axs[0].set_ylabel('Y-label')
axs[0].set_xlabel('X-label')

axs[1].set_title('title is here')
axs[1].set_ylabel('Y-label')
axs[1].set_xlabel('X-label')
fig.suptitle('BIG TITLE')
```
